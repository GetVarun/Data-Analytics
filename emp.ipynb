# Importing libraries

```python
import numpy as np 
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt 
```

# Reading file

```python
df=pd.read_csv('emp.csv')
```

```python
df.columns = df.columns.str.lower()
```

# Feature Engeneering

```python
df['salary_category'] = pd.cut(df['monthly_salary'], bins=[3500,5500,7500,9000], labels=['Low', 'Medium', 'High'], right=False)
```

```python
df['work_hours_category'] = pd.cut(df['work_hours_per_week'], bins=[30,40,50,60], labels=['Part-time','Full-Time','Over-Time'], right=False)
```

# Data analysis

1.  Average Monthly Salary by Department

```python
df.groupby('department')['monthly_salary'].mean().sort_values(ascending=False)
```

```python
df['job_title'].value_counts()
```

3. Average Monthly Salary by Gender

```python
df.groupby('gender')['monthly_salary'].mean().sort_values(ascending=False)
```

4. Correlation between Years_At_Company and Monthly_Salary

```python
df['years_at_company'].corr(df['monthly_salary'])
```

5. Average Performance Score by Job Title

6. Resignation Rate

# Data Visulization

1. Employee count by department

```python
sns.countplot(data=df, x='department', order=df['department'].value_counts().index)
plt.title('Employee Count by Department')
plt.xlabel('Department')
plt.ylabel('Number of Employees')
plt.xticks(rotation=45, ha='right')
plt.show()
```

Observation: This bar chart visually represents the distribution of employees across different departments. It quickly identifies departments with the highest and lowest headcounts, offering insights into workforce distribution and departmental size within the organization.

```python
avg_sal_job = df.groupby('job_title')['monthly_salary'].mean().sort_values(ascending=False)
```

2. Average Monthly Salary by Job Title

```python
sns.barplot(x=avg_sal_job.index, y=avg_sal_job.values, palette='mako')
plt.title('Average Monthly Salary by Job Title')
plt.xlabel('Job Title')
plt.ylabel('Average Monthly Salary')
plt.xticks(rotation=45, ha='right')
plt.show()
```

Observation: This bar chart showcases the average monthly salary for each distinct job title. It's useful for understanding the compensation hierarchy, highlighting which job roles are typically compensated at higher average rates and which are at lower rates.

3. Monthly Salary Distribution

```python
sns.histplot(df['monthly_salary'], bins=15, kde=True, color='red')
plt.title('Monthly Salary Distribution')
plt.xlabel('Monthly Salary')
plt.ylabel('Frequency')
plt.show()
```

Observation: The histogram illustrates the frequency distribution of monthly salaries. It helps in identifying the most common salary ranges among employees, whether the salaries are skewed towards lower or higher values, and if there are any significant clusters or gaps in the compensation structure.

4. Monthly Salary vs. Years at Company

```python
sns.scatterplot(data=df, x='years_at_company', y='monthly_salary', hue='job_title', size='monthly_salary', sizes=(20, 400), alpha=0.7, palette='Spectral')
plt.title('Monthly Salary vs. Years at Company by Job Title')
plt.xlabel('Years at Company')
plt.ylabel('Monthly Salary')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.)
plt.show()
```

Observation: This scatter plot explores the relationship between an employee's tenure at the company and their monthly salary, with individual points colored by job title. It can reveal:
Salary Progression: Whether there's a general trend of increasing salary with more years of service.

Job Role Influence: How different job titles impact this relationship, showing if certain roles offer faster salary growth over time or maintain higher salary levels regardless of tenure.

Compensation Consistency: Identification of any employees whose salary-to-experience ratio deviates significantly from the norm for their job title.

5. Employee Distribution by Salary Category

```python
emp_count_salary_cat = df['salary_category'].value_counts()
```

```python
plt.pie(emp_count_salary_cat, labels=emp_count_salary_cat.index, autopct='%1.1f%%', startangle=90, colors=sns.color_palette('pastel'))
plt.title('Employee Distribution by Salary Category')
plt.axis('equal') # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

Observation: The pie chart provides a clear proportional breakdown of employees across the defined 'Low', 'Medium', and 'High' salary categories. This gives a quick, high-level overview of the company's salary structure and where the majority of its workforce falls within these compensation brackets.

6. Resignation Status

```python
sns.countplot(data=df, x='resigned', palette='coolwarm')
plt.title('Employee Resignation Status')
plt.xlabel('Resigned (0=No, 1=Yes)')
plt.ylabel('Number of Employees')
plt.show()
```

Observation: This count plot (or a similar visualization) shows the number of employees who have resigned versus those who are still active. It offers a direct indication of employee turnover, which is crucial for assessing workforce stability and potential retention challenges.

7. Performance Score Distribution

```python
sns.histplot(df['performance_score'], bins=10, kde=True, color='lightgreen')
plt.title('Performance Score Distribution')
plt.xlabel('Performance Score')
plt.ylabel('Frequency')
plt.show()
```

Observation: The histogram of performance scores illustrates how employee performance ratings are distributed. This can indicate whether most employees are clustered around an average performance level, or if there's a significant number of high performers, low performers, or a bimodal distribution, reflecting the effectiveness of performance management.

8. Average Work Hours Per Week by Department

```python
avg_hours_dept = df.groupby('department')['work_hours_per_week'].mean().sort_values(ascending=False)
```

```python
sns.barplot(x=avg_hours_dept.index, y=avg_hours_dept.values, palette='crest')
plt.title('Average Work Hours Per Week by Department')
plt.xlabel('Department')
plt.ylabel('Average Work Hours')
plt.xticks(rotation=45, ha='right')
plt.show()
```

Observation: This bar chart compares the average weekly work hours across different departments. It can highlight departments that potentially have heavier workloads or different work-life balance dynamics, which might warrant further investigation for resource allocation or employee well-being initiatives.
