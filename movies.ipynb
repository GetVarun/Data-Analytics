# Data cleaning

```python
df = df.drop('ID', axis=1)
```

Converting columns to lowercase

```python
df.columns = df.columns.str.lower().str.replace(' ', '_',regex=False)
```

Converting Rotten tomatoes column to numeric by removing % 

```python
df['rotten_tomatoes'] = df['rotten_tomatoes'].astype(str).str.replace('%', '',regex=False)
df['rotten_tomatoes'] = pd.to_numeric(df['rotten_tomatoes'],errors='coerce')
```

Converting age to numerinc by removing +

```python
df['age'] = df['age'].astype(str).str.replace('+', '',regex=False)
df['age'] = pd.to_numeric(df['age'],errors='coerce')
```

Filling the null values in numeric by aggrigate function mean 

```python
df['year'] = df['year'].fillna(df['year'].mean())
df['imdb'] = df['imdb'].fillna(df['imdb'].mean())
df['rotten_tomatoes'] = df['rotten_tomatoes'].fillna(df['rotten_tomatoes'].mean())
df['runtime'] = df['runtime'].fillna(df['runtime'].mean())
df['age'] = df['age'].fillna(df['age'].mean())
```

Rounding off the columns Imdb,rotton tomatoes, runtimes,age by 2 points

```python
df['imdb'] = df['imdb'].round(2)
df['rotten_tomatoes'] = df['rotten_tomatoes'].round(2)
df['runtime'] = df['runtime'].round(2)
df['age'] = df['age'].round(2)
```

Filling the null values in non numeric columns by 'unknown'

```python
df['directors'] = df['directors'].fillna('Unknown')
df['genres'] = df['genres'].fillna('Unknown')
df['country'] = df['country'].fillna('Unknown')
df['language'] = df['language'].fillna('Unknown')
df['type'] = df['type'].fillna('Unknown')
```

# Feature Engineering 

```python
df['decade'] = (df['year'] // 10 * 10).astype(int)
```

```python
df['imdb_category'] = pd.cut(df['imdb'], bins=[0, 5.0, 7.5, 10.0], labels=['Low', 'Medium', 'High'], right=True)
```

# Data Visulization

Distribution of IMDb Scores

```python
sns.histplot(df['imdb'], bins=15, kde=True, color='skyblue')
plt.title('Distribution of IMDb Scores')
plt.xlabel('IMDb Score')
plt.ylabel('Count')
plt.show()
```

Observation: This histogram shows how IMDb scores are distributed. You can observe the most frequent score ranges (e.g., if most movies fall between 6.0 and 8.0) and identify any skewness (e.g., if there are more highly-rated movies or lower-rated ones). The KDE (Kernel Density Estimate) line provides a smoothed representation of this distribution.

Distribution of Rotten Tomatoes Scores

```python
sns.histplot(df['rotten_tomatoes'], bins=15, kde=True, color='lightcoral')
plt.title('Distribution of Rotten Tomatoes Scores')
plt.xlabel('Rotten Tomatoes Score (0-1)')
plt.ylabel('Count')
plt.show()
```

Observation: Similar to the IMDb score distribution, this histogram reveals the frequency of Rotten Tomatoes scores (on a 0-1 scale). It helps in understanding if a large number of movies are rated 'Fresh' (closer to 1.0), 'Rotten' (closer to 0), or if there's a more even spread, indicating the critical reception landscape of the dataset.

Number of Movies per Decade

```python
decade_counts = df['decade'].value_counts().sort_index()
```

```python
sns.barplot(x=decade_counts.index, y=decade_counts.values, palette='viridis')
plt.title('Number of Movies Released per Decade')
plt.xlabel('Decade')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.show()
```

Observation: This bar chart visualizes movie production trends over time. You can clearly see which decades had the highest or lowest number of movie releases, potentially reflecting industry growth, historical events, or changes in film production.

Top 10 Most Frequent Genres

```python
top_genres = df['genres'].value_counts().head(10)
```

```python
sns.barplot(x=top_genres.index, y=top_genres.values, palette='magma')
plt.title('Top 10 Most Frequent Primary Genres')
plt.xlabel('Primary Genre')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.show()
```

Observation: This chart highlights the most common movie genres in your dataset. It quickly identifies which genres are most prevalent, giving an insight into the dominant types of content within your movie collection.

Movie Count by Streaming Platform

```python
platform_counts = df[['netflix', 'hulu', 'prime_video', 'disney+']].sum()
```

```python
sns.barplot(x=platform_counts.index, y=platform_counts.values, palette='deep')
plt.title('Movie Count by Streaming Platform')
plt.xlabel('Streaming Platform')
plt.ylabel('Number of Movies')
plt.xticks(rotation=45, ha='right')
plt.show()
```

Observation: This bar chart compares the number of movies available on Netflix, Hulu, Prime Video, and Disney+. It helps in understanding which streaming platforms host the most movies from your dataset, indicating content dominance or licensing strategies.

 Relationship between IMDb and Rotten Tomatoes Scores (Scatter Plot)

```python
sns.scatterplot(data=df, x='imdb', y='rotten_tomatoes', alpha=0.6, color='darkblue', hue='imdb_category')
plt.title('IMDb vs. Rotten Tomatoes Scores')
plt.xlabel('IMDb Score')
plt.ylabel('Rotten Tomatoes Score (0-1)')
plt.legend(title='IMDb Category')
plt.show()
```

Observation: This scatter plot illustrates the correlation (or lack thereof) between IMDb user scores and Rotten Tomatoes critic scores.
If points largely follow a diagonal line, it suggests a strong positive correlation (critics and audiences generally agree).
If points are scattered, it suggests less agreement.
The imdb_category coloring helps visualize if, for example, 'High' IMDb scores consistently correspond to 'Fresh' Rotten Tomatoes scores. This plot can reveal instances where critics and audiences diverge significantly in their opinions.
